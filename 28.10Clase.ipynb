```python
from PIL import Image

image = Image.open('gato.jpg')
print(image.format)
print(image.size)
print(image.mode)

display(image)
```

```python
image_np=np.asarray(image)
print(image_np.shape)

#1193 columnas por 870 filas por 3 canales, esta es la dimension de la imagen

print(image_np.reshape(1193*870,3))

print(image_np.reshape(image_np.shape[0]*image_np.shape[1],image_np.shape[2]).shape)

print(np.unique(image_np.reshape(image_np.shape[0]*image_np.shape[1],image_np.shape[2]), axis=0).shape)

unics, freqs = np.unique(image_np.reshape(image_np.shape[0]*image_np.shape[1],image_np.shape[2]), axis=0, return_counts=True)

unics[np.argmax(freqs)]
```

```python
rgbcorrection = np.array([0,0,255])

data = np.asarray(image)
print(data.shape)
data2 = np.dot(data,rgbcorrection)

print(data2.shape)
img = Image.fromarray(np.uint8(data2))
display(img)


# as a plot
import matplotlib.pyplot as plt
plt.imshow(data2, cmap = plt.get_cmap(name = 'blue'))
plt.show()
```

```python
temperatura = np.random.randint(-10,43,1000)
```

```python
temperatura[temperatura>33]
```

```python
def isHot(grados):
    if grados>33 and grados<=40:
        return True

def isHot(grados):
    return grados>33 and grados<=40

fHot = np.vectorize(isHot)
print(fHot(temperatura))
```

```python
temperatura[fHot(temperatura)]
```
