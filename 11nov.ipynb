```python
import pandas as pd
import random

random.seed(0)

nombres = [f'Municipio{i}' for i in range(1, 11)]

data_municipios = {
    'Nombre': nombres,
    'Cdigo Postal': [random.randint(10000, 99999) for _ in range(10)],
    'Poblacin': [random.randint(1000, 50000) for _ in range(10)]
}

df_municipios = pd.DataFrame(data_municipios)

data_ayudas = {
    'Nombre':  [f'Municipio{i}' for i in range(1, 11)],
    'Ayuda Econmica (en euros)': [random.randint(1000, 10000) for _ in range(10)],
    'Nmero de Beneficiarios': [random.randint(10, 100) for _ in range(10)]
}

df_ayudas = pd.DataFrame(data_ayudas)

print("Dataframe de Municipios:")
print(df_municipios)

print("\nDataframe de Ayudas:")
print(df_ayudas)
```

```python
df_nou = df_municipios.merge(df_ayudas, left_on='Nombre', right_on='Nombre')
df_nou["Ayuda por persona"] = df_nou["Ayuda Econmica (en euros)"] / df_nou["Poblacin"]
df_nou
```
