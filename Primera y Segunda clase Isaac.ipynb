```python
# Daily data to Monthly
monthly_avg = df[["temperaturemin", "temperaturemax"]].resample("ME").mean() #resample= compactacion de todos los datos a un punto temporal-> ME=month endly, sumo todos los valores y eleboro la media
weekly_avg = df[["temperaturemin", "temperaturemax"]].resample("W").mean()
quarterly_avg = df[["temperaturemin", "temperaturemax"]].resample("QE").mean()
monthly_stats = df[["temperaturemax", "precipitation"]].resample("ME").agg({
    "temperaturemax": ["mean", "max", "min"],
    "precipitation": "sum"
})
print(monthly_avg)
print("--"*5)
print(monthly_stats)
```

```python
sample_df = df.loc["2018-01-01":"2018-01-31", ["temperaturemax", "precipitation"]].copy()
sample_df.head() #preparamos una muestra mas pequeña
```

```python
hourly_ffill = sample_df.resample("h").ffill()
print(hourly_ffill.head(27)) #en la muestra real tenemos 28,2 en el dia 0 y 32 en el dia 1, como no tengo la temperatura por hora la replica, a toda hora sera el 28.2 y asi ocurre en el dia 2
```

```python
hourly_interp = sample_df.resample("h").interpolate(method="linear")
print(hourly_interp.head(24))
```

```python
df["temp_max_7d_avg"] = df["temperaturemax"].rolling(window=7).mean()
print(df[["temperaturemax", "temp_max_7d_avg"]].head(10))
```

```python
rolling_stats = df["temperaturemax"].rolling(window=7).agg({
    "mean": "mean",
    "std": "std",
    "min": "min",
    "max": "max"
})
print(rolling_stats.head(10))
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
samples=5
df= pd.DataFrame(
    {
        "Municipio":np.repeat(["muni%i"%i for i in range(samples)],3) ,
        "Categoria"   :["Inscritos","Censo","Población"]*(samples),
        "Values"   : np.random.randint(1,10,samples*3)
    })
    
    
print(df)
```

```python
# indexcolumn, Grouper, array, or list of the previous
# Keys to group by on the pivot table index. If a list is passed, it can contain any of the other types (except list). 
# If an array is passed, it must be the same length as the data and will be used in the same manner as column values.
pd.pivot_table(df, index=['Categoria'],values="Values") #qui hay dos cosas necesarias, cual seran los nuevos valores, cual sera el nuevo indice y cual seran las nuevas columnas
#lo que obtengo es la media porque no le indico de la categoria que poner
```

```python
# Multi_index 
sales_data = pd.DataFrame({
    'Region': ['North', 'North', 'North', 'South', 'South', 'South', 'East', 'East', 'East'],
    'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Quarter': ['Q1', 'Q1', 'Q1', 'Q2', 'Q2', 'Q2', 'Q3', 'Q3', 'Q3'],
    'Sales': [100, 150, 200, 120, 160, 210, 110, 170, 190],
    'Units': [10, 15, 20, 12, 16, 21, 11, 17, 19]
})

print(sales_data)
```

#¿Que region vende mas, que Q vende mas, que producto se vende mas?

```python
pivot_multi_index1 = pd.pivot_table(sales_data,
                                   index=['Region', 'Product'],
                                   columns='Quarter',
                                   values='Sales')
print(pivot_multi_index1) #vamos a crear un multiindice
```

```python
# Incomplete data
pivot_multi_index = pd.pivot_table(sales_data,
                                   index=['Region', 'Product'],
                                   columns='Quarter',
                                   values='Sales',
                                   fill_value=0)
print(pivot_multi_index) #Le agrego que donde diga NaN me ponga un cero
```

```python
pivot_multi_index1.mean()
```

```python
pivot_multi_index.mean()
```
