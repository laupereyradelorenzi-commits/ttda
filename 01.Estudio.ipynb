```python
def es_bisiesto(anio):
    """Devuelve True si el a√±o es bisiesto, False en caso contrario."""
    if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0):
        return True
    else:
        return False

# Comprobaci√≥n de la funci√≥n
anios_prueba = [2024, 2023, 1900, 2000, 2100]

for anio in anios_prueba:
    print(f"{anio}: {'Bisiesto' if es_bisiesto(anio) else 'No bisiesto'}")
```

```python
import math

def ecuacion_segundo_grado(a, b, c):
    """Devuelve las soluciones reales o complejas de ax¬≤ + bx + c = 0."""
    discriminante = b**2 - 4*a*c

    if a == 0:
        raise ValueError("No es una ecuaci√≥n de segundo grado (a no puede ser 0).")

    if discriminante > 0:
        x1 = (-b + math.sqrt(discriminante)) / (2 * a)
        x2 = (-b - math.sqrt(discriminante)) / (2 * a)
        return x1, x2
    elif discriminante == 0:
        x = -b / (2 * a)
        return x,
    else:
        # Soluciones complejas
        parte_real = -b / (2 * a)
        parte_imaginaria = math.sqrt(-discriminante) / (2 * a)
        x1 = complex(parte_real, parte_imaginaria)
        x2 = complex(parte_real, -parte_imaginaria)
        return x1, x2

# üîç Ejemplo de comprobaci√≥n
a, b, c = 1, 10, 1  # ecuaci√≥n: x¬≤ + 10x + 1 = 0
soluciones = ecuacion_segundo_grado(a, b, c)
print("Soluciones:", soluciones)
```

```python
def hora_valida(horas, minutos, segundos):
    if 0 <= horas <= 23 and 0 <= minutos <= 59 and 0 <= segundos <= 59:
        print("La hora es v√°lida.")
    else:
        print("La hora NO es v√°lida.")

hora_valida(12, 45, 30)  # ‚úÖ v√°lida
hora_valida(25, 10, 0)   # ‚ùå no v√°lida
```

```python
n = int(input("Ingresa un n√∫mero entero: "))

# Calcular la suma de los divisores propios
suma_divisores = 0
for i in range(1, n):
    if n % i == 0:
        suma_divisores += i

# Clasificar el n√∫mero
if suma_divisores == n:
    print(f"{n} es un n√∫mero PERFECTO.")
elif suma_divisores < n:
    print(f"{n} es un n√∫mero DEFICIENTE.")
else:
    print(f"{n} es un n√∫mero ABUNDANTE.")
```
