```python
df2 = pd.DataFrame([(1, 145.0, 3000.1), #No 2
                   ( 3, 129.0, 3000.0), # Multiples 3
                   ( 3, 159.0, 4000.0),
                   ( 3, 109.0, 5000.0),
                   ( 4, 198.1, 4020.2),
                   ( 5, 200.0, 5000.2)], #a new one
           columns=('DNI', 'Weight', 'Salary')
                 )
df2
```

```python
df1.merge(df2, left_on='ID', right_on='DNI')
```

```python

#Si una fila en df1 no tiene una coincidencia en df2, las columnas de df2 en esa fila se rellenarán con NaN (Not a Number, valor faltante).
```

```python
df1.merge(df2, left_on='ID', right_on='DNI',how="right")
```

```python
 df1.merge(df2, left_on='ID', right_on='DNI',how="inner")#El DataFrame final contendrá solo los registros que tienen un valor coincidente en la columna ID de df1 y la columna DNI de df2. Es la forma más restrictiva de unión.
```

```python
df1.merge(df2,how="cross") #todas las filas del DF1 mezclalas con las filas del DF 2
```
