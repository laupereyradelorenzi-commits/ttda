```python
estudiantes = pd.DataFrame({
"id": [1, 2, 3, 4],
"nombre": ["Ana", "Luis", "Marta", "Jorge"]
})
estudiantes
```

```python
df_merge_interno = pd.merge(estudiantes, notas, on="id", how="inner")
df_merge_interno
```

```python
#Se obtienen tres filas porque el DF_notas contiene solo 3 id (1,2,5) y el DF_estudiantes (1,2,3,4), por ello la "union" solo se hace de los que son compatibles.
#En el de estudiantes son dos ID uno corresponde a 1 y otro a 2, en cambio en el de notas hay uno que corresponde a 1 y dos con 2 por ello el resultado son 3.
```

```python
df_left_merge = pd.merge(estudiantes, notas, on="id", how="left")
df_left_merge
```

```python
#Dado que el left merge utiliza estudiantes como tabla izquierda, solo mantiene los IDs que están presentes en estudiantes.
#En el df_estudiantes hay 4 ID en total, en cambio en el df_notas solo 3 (1,2,5), esta union hace que este dos veces el 2 de alumnos porque hay dos veces Id 2 de nota.
#En conclusiòn, se pierde el ID 5 de nota
```

```python
#EJERCICIO 5 Agregaciones y operaciones con groupby
```

```python
ventas = pd.DataFrame({
"region": ["Norte","Norte","Sur","Este","Sur","Este","Oeste",
"Oeste","Norte","Sur"],
"Vendedor": ["Ana","Luis","Ana","Jorge","Luis","Ana","Marta",
"Luis","Ana","Jorge"],
"ventas": [1200, 800, 950, 1100, 400, 1600, 700, 450, 895, 900],
"producto": ["A","A","B","A","B","A","B","A","B","B"]
})
ventas
```

```python
totales_region = ventas.groupby("region")["ventas"].sum().reset_index()
totales_region
```

```python
#Ejercicio 5.b
```

```python
ventas.groupby("Vendedor")["ventas"].agg(
    media="mean",
    maximo="max",
    minimo="min"
)
```

```python
#EJERCICIO 6. Gestión de datos ausentes.
```

```python
estudiantes = pd.DataFrame({
"id": [1, 2, 3, 4],
"nombre": ["Ana", "Luis", "Marta", "Jorge"]
})

notas = pd.DataFrame({
"id": [1, 2, 2, 5],
"asignatura": ["Matematicas", "Historia", "Quımica", "Biologia"],
"nota": [8.5, 7.0, 6.5, 9.0]
})
```

```python
df_merged = pd.merge(estudiantes, notas, on='id', how='left')
df_merged
```

```python
nans_por_columna = df_merged.isna().sum().reset_index(name="NaN")
nans_por_columna
```

```python
#EJERCICIO 6.b
```

```python
df_reemplazo = df_merged.fillna({'asignatura': 0, 'nota': 0})
df_reemplazo
```
