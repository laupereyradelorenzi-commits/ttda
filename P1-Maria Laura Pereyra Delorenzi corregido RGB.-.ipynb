```python
import numpy as np
```

```python
from PIL import Image
```

```python
#Ejercicio 1
image = Image.open('palma.jpg')
print(image.format)
print(image.size)
print(image.mode)

display(image)
```

```python
#Ejercicio 2
data = np.asarray(image)
print(np.mean(data))
print(np.max(data))
print(np.min(data))
```

```python
print(data.shape)
```

```python
print(np.mean(data[:,:,0]))
print(np.mean(data[:,:,1]))
print(np.mean(data[:,:,2]))
print(np.max(data[:,:,0]))
print(np.max(data[:,:,1]))
print(np.max(data[:,:,2]))
print(np.min(data[:,:,0]))
print(np.min(data[:,:,1]))
print(np.min(data[:,:,2]))
```

```python
rojo = data[:, :, 0]
verde = data[:, :, 1]
azul = data[:, :, 2]
print("R")
print(f"Mnimo: {np.min(rojo)}")
print(f"Mximo: {np.max(rojo)}")
print(f"Media: {np.mean(rojo)}")

print("G")
print(f"Mnimo: {np.min(verde)}")
print(f"Mximo: {np.max(verde)}")
print(f"Media: {np.mean(verde)}")

print("B")
print(f"Mnimo: {np.min(azul)}")
print(f"Mximo: {np.max(azul)}")
print(f"Media: {np.mean(azul)}")
```

```python
#Ejercicio 3
data = np.asarray(image)
data2 = data.copy()
for x in range(data2.shape[0]):
     for y in range(data2.shape[1]):
         data2[x,y] = np.repeat(data[x,y].mean(),3)

img = Image.fromarray(data2, 'RGB')
display(img)
```
