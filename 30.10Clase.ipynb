```python
a = np.arange(10)
b = np.arange(5,15)
print(a)
print(b)
print("-"*10)
print(np.in1d(a,b))
print(np.intersect1d(a,b))
print("-"*10)
print(np.setdiff1d(a,b))
print(np.setdiff1d(b,a))
print("-"*10)
print(np.union1d(a,b))
```

```python
x = np.array([0, 0,2, 2, 3, 4, 54, 6, 7, 80, 9])
print(np.diff(x))
print(np.sign(np.diff(x)))
print(np.argwhere (np.sign(np.diff(x))==-1))# en que posiciones esta el menos 1
print(x[1:][np.argwhere (np.sign(np.diff(x))==-1)-1])
```
